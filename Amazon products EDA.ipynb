### Task

In this,I will use Amazon Products Reviews dataset of about 34000 customer reviews, obtained from Kaggle website, and will try to explore it and visualize it's numerical and categorical features as much as I can, using Plotly’s Python graphing library.

I will also make visualisations of single and two-worded phrases used by customers in the reviews by using n-gram feature which is a contiguous sequence of n items from a given sample of text or speech.

```python
import pandas as pd
```

```python
df=pd.read_csv('amazon_products_reviews.csv',low_memory=False) #'low_memory= False' because guessing dtypes for each column is very 
                                                                 #memory demanding as the dataset contains around 34000 rows 
df.tail()  
```

#### Data preprocessing

```python
#dropping unnecessary columns from which I would not get required information
df.drop(['reviews.date','reviews.dateAdded','reviews.dateSeen','reviews.didPurchase','reviews.id','reviews.numHelpful','reviews.userCity','reviews.userProvince'],axis=1, inplace=True) 

#checking if any review text is nill or not
df = df[~df['reviews.text'].isnull()] 
```

```python
df.describe()   #-- shows the descriptive analysis of numerical features mentioned in the dataset
```

```python
 df['categories']
```

```python
# check weather dataset contain any null values
df.isnull().sum()

# replace them with '0'
df.fillna(0,inplace=True)
df.head()
```

#### Using TextBlob to calculate sentiment polarity which lies in the range of [-1,1] where 1 means positive sentiment and -1 means a negative sentiment.

```python
from textblob import TextBlob 
```

```python
df['polarity'] = df['reviews.text'].map(lambda text: TextBlob(text).sentiment.polarity)
df['review_len'] = df['reviews.text'].astype(str).apply(len)
df['word_count'] = df['reviews.text'].apply(lambda x: len(str(x).split())) 
```

```python
df.columns
```

##### To preview whether the sentiment polarity score works, we randomly select 3 reviews with the highest sentiment polarity score==1 :

```python
print('3 random reviews with the highest positive sentiment polarity:- \n')
ps = df.loc[df.polarity == 1, ['reviews.text']].sample(3).values
for i in ps:
    print(i[0])  
```

#### Then randomly select 3 reviews with the neutral sentiment polarity score i.e. 0 :

```python
print('3 random reviews with the most neutral sentiment(zero) polarity: \n')
ps = df.loc[df.polarity == 0, ['reviews.text']].sample(3).values
for i in ps:
    print(i[0])
```

#### And at last, randomly select 3 reviews with the most negative sentiment polarity score i.e. -1

```python
print('3 random reviews with the most negative polarity: \n')
ps = df.loc[df.polarity ==-1, ['reviews.text']].sample(3).values
for i in ps:
    print(i[0]) 
```

#### So, all of them did work!

### Univariate visualization with Plotly

### Now, we are going to see which words are written most frequently in "reviews" by the customers before and after removing stopwords. 

#### Stopwords- these are the English words which does not add much meaning to a sentence -set of words 

#####  I come to “reviews.text” feature and before exploring this I need to extract "N-Gram features". 

N-grams are used to describe the number of words used as observation points, 
For e.g., Unigram means singly-worded, bigram means 2-worded phrase. 
In order to do this, I will use scikit-learn’s CountVectorizer function.

 It's interesting to compare unigrams before and after removing stop words.

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
def top_n_words(corpus, n=None):
    vec = CountVectorizer().fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = top_n_words(df['reviews.text'], 15)
# for word, freq in common_words:
#     print(word, freq)
df1 = pd.DataFrame(common_words, columns = ['reviews.text' ,'count'])
df1.groupby('reviews.text').sum()['count'].sort_values(ascending=False).iplot(kind='bar',
    yTitle='Count', linecolor='black', title='Top 15 words in review text before removing stop words(unigrams)')
```

```python
def top_n_words(corpus, n=None):
    vec = CountVectorizer(stop_words = 'english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = top_n_words(df['reviews.text'], 15)
# for word, freq in common_words:
#     print(word, freq)
df2 = pd.DataFrame(common_words, columns = ['reviews.text' , 'count'])
df2.groupby('reviews.text').sum()['count'].sort_values(ascending=False).iplot(
kind='bar', yTitle='Count', linecolor='black', title='Top 15 words in review text after removing stop words(unigrams)')
```

```python
def top_n_bigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(2, 2)).fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = top_n_bigram(df['reviews.text'], 15)

df3 = pd.DataFrame(common_words, columns = ['reviews.text' , 'count'])
df3.groupby('reviews.text').sum()['count'].sort_values(ascending=False).iplot(kind='bar', yTitle='Count', 
                     linecolor='black', title='Top 15 words in review text before removing stop words(bigrams)')
```

```python
def top_n_bigram(corpus, n=None):
    vec = CountVectorizer(ngram_range=(2, 2), stop_words='english').fit(corpus)
    bag_of_words = vec.transform(corpus)
    sum_words = bag_of_words.sum(axis=0) 
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
    return words_freq[:n]
common_words = top_n_bigram(df['reviews.text'], 15)

df4 = pd.DataFrame(common_words, columns = ['reviews.text' , 'count'])
df4.groupby('reviews.text').sum()['count'].sort_values(ascending=False).iplot(
kind='bar', yTitle='Count', linecolor='black', title='Top 15 words in review text after removing stop words(bigrams)')
```
